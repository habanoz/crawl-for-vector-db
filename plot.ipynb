```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.title('Number of words by number of links (min_char = 0)')

plt.plot([1,4,7],[1364, 5125, 9125], 'o-r', label='BlockParser')
plt.plot([1,4,7],[1575, 6118, 11581],  'o-g', label='WebBasedLoader')
plt.ylabel('Total words')
plt.xlabel('#Links')
plt.xticks([1,4,7])
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Number of words by min char limit (#links = 4)')

plt.plot([0,20,50],[5125, 4999, 4380], 'o-r', label='BlockParser')
plt.plot([0,20,50],[6118, 5929, 6117],  'o-g', label='WebBasedLoader')
plt.ylabel('Total words')
plt.xlabel('min char limit')
plt.xticks([0,20,50])
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Average #words per docs by min char limit (#links = 4)')

plt.plot([0,20,50],[16.5,18.7,25.9], 'o-r', label='BlockParser')
plt.plot([0,20,50],[278,282.3,278], 'o-g', label='WebBasedLoader')
plt.ylabel('Average #words (Log scale)')
plt.xlabel('min char limit')
plt.yscale("log")
plt.grid()
plt.legend()
```
